```python
from Bio import SeqIO

# *********************Question 1: Biopython and Loops*********************
# Initialize counter for sequences containing "ATGCG"
count_atgcg = 0

# Read the FASTA file using Biopython
for record in SeqIO.parse("genes.fa", "fasta"):
    sequence = str(record.seq).upper()

    # Check if the sequence contains the substring "ATGCG"
    if "ATGCG" in sequence:
        count_atgcg += 1

# Output the count of sequences containing "ATGCG"
print(f"Number of sequences containing 'ATGCG': {count_atgcg}")


# *********************Question 2: Biopython and Conditional if-elif-else*********************
with open("gene_length_report.txt", "w") as output_file:

    # Read the FASTA file using Biopython
    for record in SeqIO.parse("genes.fa", "fasta"):
        gene_name = record.id
        sequence_length = len(record.seq)

        # Classify gene sequence based on its length
        if sequence_length < 250:
            classification = "Very Short"
        elif 250 <= sequence_length <= 800:
            classification = "Moderate"
        else:
            classification = "Long"

        # Write result to the output file
        output_file.write(f"{gene_name}\t{classification}\n")

print("Classification results written to 'gene_length_report.txt'")


# *********************Question 3: Biopython*********************
total_genes = 0
shortest_gene_length = float("inf")
shortest_gene_name = None

# Read the FASTA file using Biopython
for record in SeqIO.parse("genes.fa", "fasta"):
    total_genes += 1
    gene_length = len(record.seq)

    # Check for shortest gene
    if gene_length < shortest_gene_length:
        shortest_gene_length = gene_length
        shortest_gene_name = record.id

print(f"Total number of genes: {total_genes}")
print(f"Shortest gene: {shortest_gene_name} (Length: {shortest_gene_length})")


# *********************Question 4: Biopython and Dictionary (GC Content)*********************
def calculate_gc(sequence):
    gc_count = sequence.count("G") + sequence.count("C")
    return (gc_count / len(sequence)) * 100 if len(sequence) > 0 else 0


gene_gc_dict = {}

for record in SeqIO.parse("genes.fa", "fasta"):
    gene_name = record.id
    sequence = str(record.seq).upper()

    gene_gc_dict[gene_name] = calculate_gc(sequence)

print("\nGene GC Content Dictionary:")
for gene, gc in gene_gc_dict.items():
    print(f"{gene}: {gc:.2f}%")

# Find gene with highest GC content
highest_gc_gene = max(gene_gc_dict, key=gene_gc_dict.get)
print(f"\nGene with highest GC content: {highest_gc_gene} ({gene_gc_dict[highest_gc_gene]:.2f}%)")


# *********************Question 5: Biopython and Functions (Lowest GC Content)*********************
# Reusing calculate_gc function

lowest_gc_gene = None
lowest_gc_value = float("inf")

for record in SeqIO.parse("genes.fa", "fasta"):
    gene_sequence = str(record.seq).upper()
    gc_value = calculate_gc(gene_sequence)

    if gc_value < lowest_gc_value:
        lowest_gc_value = gc_value
        lowest_gc_gene = record.id

print(f"\nGene with lowest GC content: {lowest_gc_gene} ({lowest_gc_value:.2f}%)")
```
