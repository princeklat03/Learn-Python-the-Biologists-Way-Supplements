```python
from Bio import SeqIO
import re

file_name = "Organism_cDNA_FASTA.txt"


# Question 1
count_start_gtg = 0
genes_start_gtg = []

# Question 2
count_stop_codons = 0

# Question 3
longest_a_run = 0
longest_a_gene = None
longest_a_sequence = ""

# Question 4
count_poly_t = 0
genes_with_poly_t = []

# Question 5
motif = "AAGCTT"

count_motif_genes = 0
genes_with_motif = []


# Process FASTA file
for record in SeqIO.parse(file_name, "fasta"):

    sequence = str(record.seq).upper()
    gene_name = record.id

    # ********************************************************
    # Question 1
    # ********************************************************

    if re.search(r"^GTG", sequence):

        count_start_gtg += 1
        genes_start_gtg.append(gene_name)


    # ********************************************************
    # Question 2
    # ********************************************************

    if re.search(r"(TAA|TAG|TGA)$", sequence):

        count_stop_codons += 1


    # ********************************************************
    # Question 3
    # ********************************************************

    a_run_matches = re.findall(r"A{4,}", sequence)

    if a_run_matches:

        max_a_run = max(a_run_matches, key=len)

        if len(max_a_run) > longest_a_run:

            longest_a_run = len(max_a_run)
            longest_a_gene = gene_name
            longest_a_sequence = max_a_run


    # ********************************************************
    # Question 4
    # ********************************************************

    t_matches = re.findall(r"T{6,}", sequence)

    if t_matches:

        count_poly_t += 1

        genes_with_poly_t.append(
            f"{gene_name} ({', '.join(t_matches)})"
        )


    # ********************************************************
    # Question 5
    # ********************************************************

    motif_matches = re.findall(motif, sequence)

    if motif_matches:

        count_motif_genes += 1

        genes_with_motif.append(
            f"{gene_name} ({len(motif_matches)} occurrence(s))"
        )


# ********************************************************
# Output Results
# ********************************************************

print("Question 1:")
print(f"Genes starting with 'GTG': {count_start_gtg}")
print("Gene Names:")
print(", ".join(genes_start_gtg))


print("\nQuestion 2:")
print(f"Genes ending with a stop codon: {count_stop_codons}")


print("\nQuestion 3:")
print(f"Gene with the longest consecutive 'A' stretch: {longest_a_gene}")
print(f"Longest stretch: {longest_a_sequence}")
print(f"Length of stretch: {longest_a_run}")


print("\nQuestion 4:")
print(f"Genes with 6 or more consecutive 'T': {count_poly_t}")
print("Gene Names and Matching Stretches:")

for gene in genes_with_poly_t:
    print(gene)


print("\nQuestion 5:")
print(f"Genes containing motif '{motif}': {count_motif_genes}")
print("Gene Names and Motif Counts:")

for gene in genes_with_motif:
    print(gene)
```
