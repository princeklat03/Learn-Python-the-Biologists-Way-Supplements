```python
"""
Answer Notebook for Chapter 5
Learn Python the Biologist's Way

This notebook provides code-based solutions for the
activities and exercises presented in Chapter 5.

The solutions presented here are sample approaches
and may not be the only valid ways to solve these
problems. Readers are encouraged to explore
alternative approaches and adapt the code as needed.

The book purchase link and additional resources are
available on the GitHub repository landing page

Author: Dr. Prince Kumar Lat
"""

#Part 1
#Question 1

frag_1="GATCCATGATCGTACGTTAG"
frag_2="TTACGATCGATGCTCTGCA"
frag_3="GATTGCAGGGTTAACCG"

dna_seq=frag_2+frag_3+frag_1
print(dna_seq)


#Question 2

comp=dna_seq.replace("A", "t")
comp=comp.replace("T", "a")
comp=comp.replace("C", "g")
comp=comp.replace("G", "c")
comp=comp.upper()

rev_comp = comp[::-1]
print(rev_comp)


#Question 3

Tot=len(dna_seq)

Num_A=dna_seq.count("A")
Num_T=dna_seq.count("T")
Num_G=dna_seq.count("G")
Num_C=dna_seq.count("C")

print("A: "+"%.2f"%(Num_A/Tot*100)+'%')
print("T: "+"%.2f"%(Num_T/Tot*100)+'%')
print("G: "+"%.2f"%(Num_G/Tot*100)+'%')
print("C: "+"%.2f"%(Num_C/Tot*100)+'%')


#Part 2
#Question 1

full_seq="ATGCGTACGTAGCTAGCTAGGCTAGCTAGCTAACGATCGTAGCTAGCTAACGATCGTAGCTAGCTAGCTAGCTAGCTA"

exon_1 = full_seq[:20]
exon_2 = full_seq[30:50]
exon_3 = full_seq[60:]

coding_region=exon_1+exon_2+exon_3

print(len(coding_region))
print(coding_region)


#Question 2

coding=len(coding_region)/len(full_seq)*100
non_coding=100-coding

print("Coding: "+"%.2f"%coding+'%')
print("Non-Coding: "+"%.2f"%non_coding+'%')


#Question 3

Intron_1 = full_seq[20:30].lower()
Intron_2 = full_seq[50:60].lower()

dna=exon_1+Intron_1+exon_2+Intron_2+exon_3
print(dna)


#Optional Question 4 (Mutation Simulation)

mutated_seq = coding_region.replace("A", "T", 1)

print("Original:", coding_region)
print("Mutated:", mutated_seq)

changes = 0
for i in range(len(coding_region)):
    if coding_region[i] != mutated_seq[i]:
        changes += 1

print("Number of changes:", changes)
```
