```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Jan 7 20:19:43 2025

@author: PK Lat
"""

#Part 1
#Question 1

frag_1="GATCCATGATCGTACGTTAG"
frag_2="TTACGATCGATGCGTAC"
frag_3="CTGCAGGGTTAACCG"

# Reconstructing sequence (assumed correct order based on cut sites)
dna_seq=frag_2+frag_1+frag_3
print(dna_seq)


#Question 2 (Reverse Complement)

comp=dna_seq.replace("A", "t")
comp=comp.replace("T", "a")
comp=comp.replace("C", "g")
comp=comp.replace("G", "c")
comp=comp.upper()

rev_comp = comp[::-1]
print(rev_comp)


#Question 3 (Nucleotide Composition)

Tot=len(dna_seq)

No_A=dna_seq.count("A")
No_T=dna_seq.count("T")
No_G=dna_seq.count("G")
No_C=dna_seq.count("C")

print("A: "+"%.2f"%(No_A/Tot*100)+'%')
print("T: "+"%.2f"%(No_T/Tot*100)+'%')
print("G: "+"%.2f"%(No_G/Tot*100)+'%')
print("C: "+"%.2f"%(No_C/Tot*100)+'%')


#Part 2
#Question 4

full_seq="ATGCGTACGTAGCTAGCTAGGCTAGCTAGCTAACGATCGTAGCTAGCTAACGATCGTAGCTAGCTAGCTAGCTAGCTA"

exon_1 = full_seq[:20]
exon_2 = full_seq[30:50]
exon_3 = full_seq[60:]

coding_region=exon_1+exon_2+exon_3

print(len(coding_region))
print(coding_region)


#Question 5

coding=len(coding_region)/len(full_seq)*100
non_coding=100-coding

print("Coding: "+"%.2f"%coding+'%')
print("Non-Coding: "+"%.2f"%non_coding+'%')


#Question 6

Intron_1 = full_seq[20:30].lower()
Intron_2 = full_seq[50:60].lower()

dna=exon_1+Intron_1+exon_2+Intron_2+exon_3
print(dna)


#Optional Question 7 (Mutation Simulation)

mutated_seq = coding_region.replace("A", "T", 1)

print("Original:", coding_region)
print("Mutated:", mutated_seq)

changes = 0
for i in range(len(coding_region)):
    if coding_region[i] != mutated_seq[i]:
        changes += 1

print("Number of changes:", changes)
```
