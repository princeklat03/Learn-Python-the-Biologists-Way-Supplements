```python
"""
Answer Notebook for Chapter 6
Learn Python the Biologist's Way

This notebook provides code-based solutions for the
activities and exercises presented in Chapter 6.

The solutions presented here are sample approaches
and may not be the only valid ways to solve these
problems. Readers are encouraged to explore
alternative approaches and adapt the code as needed.

The book purchase link and additional resources are
available on the GitHub repository landing page

Author: Dr. Prince Kumar Lat
"""

# Path to the FASTA file
fasta_path = "MYC.fna"
# Output file
output_path = "MYC_summary.txt"

with open(fasta_path, "r") as fasta_file, open(output_path, "w") as output_file:
    sequence_name = ""
    sequence = ""

    for line in fasta_file:
        line = line.strip()

        if line.startswith(">"):
            if sequence != "":
                length = len(sequence)
                gc_count = sequence.count("G") + sequence.count("C")
                gc_content = (gc_count / length) * 100
                at_count = sequence.count("A") + sequence.count("T")
                at_content = (at_count / length) * 100
                output_file.write(f"Sequence: {sequence_name}\n")
                output_file.write(f"Length: {length}\n")
                output_file.write(f"GC Content: {gc_content:.2f}%\n")
                output_file.write(f"AT Content: {at_content:.2f}%\n")
                output_file.write("\n")

            sequence_name = line[1:]
            sequence = ""
        else:
            sequence = sequence + line.upper()
    if sequence != "":
        length = len(sequence)
        gc_count = sequence.count("G") + sequence.count("C")
        gc_content = (gc_count / length) * 100
        at_count = sequence.count("A") + sequence.count("T")
        at_content = (at_count / length) * 100
        output_file.write(f"Sequence: {sequence_name}\n")
        output_file.write(f"Length: {length}\n")
        output_file.write(f"GC Content: {gc_content:.2f}%\n")
        output_file.write(f"AT Content: {at_content:.2f}%\n")
```
