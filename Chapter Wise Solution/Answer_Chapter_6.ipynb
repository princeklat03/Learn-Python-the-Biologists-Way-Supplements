```python
# Path to the FASTA file
fasta_path = "MYC.fna"
# Output file
output_path = "MYC_summary.txt"
# Read FASTA file and write results
with open(fasta_path, "r") as fasta_file, open(output_path, "w") as output_file:
    sequence_name = ""
    sequence = ""

    for line in fasta_file:
        line = line.strip()

        # Check if line is a header
        if line.startswith(">"):
            # Process previous sequence before moving to next one
            if sequence != "":
                length = len(sequence)
                gc_count = sequence.count("G") + sequence.count("C")
                gc_content = (gc_count / length) * 100
                at_count = sequence.count("A") + sequence.count("T")
                at_content = (at_count / length) * 100
                output_file.write(f"Sequence: {sequence_name}\n")
                output_file.write(f"Length: {length}\n")
                output_file.write(f"GC Content: {gc_content:.2f}%\n")
                output_file.write(f"AT Content: {at_content:.2f}%\n")
                output_file.write("\n")

            # Store new sequence header
            sequence_name = line[1:]
            # Reset sequence
            sequence = ""
        else:
            # Add sequence lines together
            sequence = sequence + line.upper()
    # Process the last sequence in the file
    if sequence != "":
        length = len(sequence)
        gc_count = sequence.count("G") + sequence.count("C")
        gc_content = (gc_count / length) * 100
        at_count = sequence.count("A") + sequence.count("T")
        at_content = (at_count / length) * 100
        output_file.write(f"Sequence: {sequence_name}\n")
        output_file.write(f"Length: {length}\n")
        output_file.write(f"GC Content: {gc_content:.2f}%\n")
        output_file.write(f"AT Content: {at_content:.2f}%\n")
```
