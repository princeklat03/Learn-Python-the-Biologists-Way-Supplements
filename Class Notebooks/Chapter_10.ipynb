```python
from Bio import SeqIO
from Bio.SeqUtils import gc_fraction, molecular_weight

fasta_file = SeqIO.parse("c_elegans_100.fa", "fasta")

for fasta in fasta_file:
    seq_id = fasta.id
    seq = fasta.seq
    gc = round(gc_fraction(seq), 2)
    weight = round(molecular_weight(seq, "DNA"), 2)
    print(f"{seq_id}   GC:{gc}   MW:{weight}")
```

```python
from Bio import Entrez, Medline

# Set up
Author = "Lat PK"
database = "pubmed"
retmax = 10  # Maximum number of articles to fetch
Entrez.email = "prince@redwoodai.com" # Required by NCBI

# Step 1: Search for articles by the author
handle = Entrez.esearch(
        db=database,
        term=f"{Author} [Author]",
        retmax=retmax
)
record = Entrez.read(handle)
handle.close()

id_list = record["IdList"]
print(f"Found PubMed IDs: {id_list}")

# Step 2: Fetch the article details in Medline format
ids = ",".join(id_list)
handle = Entrez.efetch(
        db=database,
        id=ids,
        rettype="medline",
        retmode="text"
)
records = list(Medline.parse(handle))
handle.close()

# Step 3: Save abstracts to a text file
filename = (
        f"Abstracts_by_{Author.replace(' ', '_')}.txt"
)
with open(filename, "w", encoding="utf-8") as f:
    for i, rec in enumerate(records, start=1):
        title = rec.get("TI", "No Title")
        journal = rec.get("JT", "No Journal")
        year = rec.get("DP", "No Date")
        abstract = rec.get("AB", "No Abstract")

        f.write(f"Article {i}:\n")
        f.write(f"Title: {title}:\n")
        f.write(f"Journal: {journal} ({year})\n")
        f.write(f"Abstract:\n{abstract}\n")
        f.write("-" * 80 + "\n\n")

print(f"Saved {len(id_list)} abstracts to file.")
```

```python
from Bio import Entrez, Medline
# List of nucleotide accession IDs
ids = ["NM_001301717", "NM_001101", "NM_000546"]

# Fetch sequences from NCBI
handle = Entrez.efetch(
          db="nucleotide",
          id=",".join(ids),
          rettype="fasta",
          retmode="text"
)

# Write sequences to a FASTA file
with open("sequences_ids.fasta", "w") as f:
    f.write(handle.read())

handle.close()

print(f"Saved sequences for {len(ids)} IDs to ‘sequences_ids.fasta’.")
```
