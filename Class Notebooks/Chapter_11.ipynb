```python
from rdkit import Chem

# Input SMILES
smiles = "CCO"
print(f"smiles: {smiles}")

# Convert SMILES to molecular graph
mol_smiles = Chem.MolFromSmiles(smiles)
print(f"mol_smiles: {mol_smiles}")

# Recover an augmented (non-canonical) SMILES
smiles_generated = Chem.MolToSmiles(
    mol_smiles,
    doRandom=True,
    canonical=False
)
print(f"smiles_ generated: {smiles_generated}")
```

```python
from rdkit import Chem

# Input SMILES
smiles = "OCC"
print(f"smiles: {smiles}")

# Convert SMILES to molecular graph
mol_smiles = Chem.MolFromSmiles(smiles)
print(f"mol_smiles: {mol_smiles}")

# Recover canonical SMILES from the molecular object
smiles_canonical = Chem.MolToSmiles(
    mol_smiles,
    isomericSmiles=True,
    canonical=True
)
print(f"smiles_canonical: {smiles_canonical}")
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Two SMILES representations of the same molecule
augmented_smiles = "OCC"
canonical_smiles = "CCO"

# Convert SMILES string to molecular graphs
augmented_smiles_mol = Chem.MolFromSmiles(augmented_smiles)
canonical_smiles_mol = Chem.MolFromSmiles(canonical_smiles)

# Draw the molecule (augmented SMILES)
Draw.MolToImage(
    augmented_smiles_mol,
    size=(400, 400)
)
```

```python
Draw.MolToImage(
    canonical_smiles_mol,
    size=(400, 400)
)
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Complex molecule represented as a SMILES string
smiles = "C1COCCN1CCCOc2c(OC)cc3ncnc(c3c2)Nc4cc(Cl)c(F)cc4"

# Convert SMILES to molecular graph
mol_smiles = Chem.MolFromSmiles(smiles)

# Draw the molecule
Draw.MolToImage(
        mol_smiles,
        size=(600, 400)
)
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Complex molecule represented as a SMILES string
smiles = "C1COCCN1CCCOc2c(OC)cc3ncnc(c3c2)Nc4cc(Cl)c(F)cc4"

# Convert SMILES to molecular graph
mol_smiles = Chem.MolFromSmiles(smiles)

# Draw the molecule without Kekulé representation
Draw.MolToImage(
    mol_smiles,
    size=(600, 400),
    kekulize=False
)
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import rdMolDraw2D

# Complex molecule represented as a SMILES string
smiles = "C1COCCN1CCCOc2c(OC)cc3ncnc(c3c2)Nc4cc(Cl)c(F)cc4"

# Convert SMILES to molecular graph
mol_smiles = Chem.MolFromSmiles(smiles)

# Set drawing options
opts = rdMolDraw2D.MolDrawOptions()
opts.useBWAtomPalette()

# Draw the molecule without aromatic bonds preserved
img = Draw.MolToImage(
        mol_smiles,
        size=(600, 400),
        kekulize=False,
        options=opts
)

img
```

```python
from rdkit import Chem

# Acetaminophen (Paracetamol)
smiles = "CC(=O)Nc1ccc(O)cc1"
```

```python
mol = Chem.MolFromSmiles(smiles)
```

```python
inchi = Chem.MolToInchi(mol)
print(f"InChI: {inchi}")
```

```python
smarts = Chem.MolToSmarts(mol)
print(f"SMARTS: {smarts}")
```

```python
mol_from_inchi = Chem.MolFromInchi(inchi)
smiles_from_inchi = Chem.MolToSmiles(
    mol_from_inchi,
    canonical=True
)

print(f"Recovered SMILES from InChI: {smiles_from_inchi}")
```

```python
from rdkit import Chem

# Convert SMARTS back to a molecule (pattern-based)
mol_from_smarts = Chem.MolFromSmarts(smarts)

# Attempt to generate SMILES from the SMARTS-derived molecule
smiles_from_smarts = Chem.MolToSmiles(
    mol_from_smarts
)

print(f"Recovered from SMARTS: {smiles_from_smarts}")
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

# Acetaminophen SMILES
smiles = "CC(=O)Nc1ccc(O)cc1"

# Convert SMILES to molecular graph
mol = Chem.MolFromSmiles(smiles)

# Basic molecular properties
print("SMILES:", Chem.MolToSmiles(mol))
print("Molecular Weight:", Descriptors.MolWt(mol))
print("Molecular Formula:", Chem.rdMolDescriptors.CalcMolFormula(mol))

# physicochemical properties
print("LogP:", Descriptors.MolLogP(mol))
print("Number of H-bond Donors:", Descriptors.NumHDonors(mol))
print("Number of H-bond Acceptors:", Descriptors.NumHAcceptors(mol))
```
