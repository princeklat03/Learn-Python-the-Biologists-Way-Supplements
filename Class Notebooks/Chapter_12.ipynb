```python
dict3 = {
    "gene_name": "Tp53",
    "location": "17p13.1",
    "disease": "Li-Fraumeni syndrome",
    "expression": 150
}
```

```python
# value_deleted = dict3_copy.pop("<key>")
```

```python
print("Before:", dict3_copy)
value_deleted = dict3_copy.pop("location")
print("Deleted Value:", value_deleted)
print("After:", dict3_copy)
```
