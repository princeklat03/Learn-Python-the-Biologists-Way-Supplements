```python
def do_reverse_complement(input_dna):
    reverse_comp_dna = input_dna.replace("A", "t")
    reverse_comp_dna = reverse_comp_dna.replace("T", "a")
    reverse_comp_dna = reverse_comp_dna.replace("G", "c")
    reverse_comp_dna = reverse_comp_dna.replace("C", "g")

    reverse_comp_dna = reverse_comp_dna.upper()
    reverse_comp_dna = reverse_comp_dna[::-1]

    return reverse_comp_dna
```

```python
dna1 = "ATTGGGCC"
dna2 = "ATCGTTAG"

result1 = do_reverse_complement(dna1)
result2 = do_reverse_complement(dna2)

print(result1)
print(result2)
```

```python
import random

def generate_random_dna():
    nucleotides = ["A", "T", "G", "C"]
    dna_sequence = "".join(random.choice(nucleotides) for i in range(10))
    return dna_sequence
```

```python
def calculate_and_compare_gc(dna_seq1, dna_seq2, decimals):
    gc1 = round(gc_fraction(dna_seq1), decimals)
    gc2 = round(gc_fraction(dna_seq2), decimals)

    if gc1 > gc2:
        comparison = "greater"
    else:
        comparison = "less"

    return gc1, gc2, comparison
```

```python
dna1 = generate_random_dna()
dna2 = generate_random_dna()

gc1, gc2, result = calculate_and_compare_gc(dna1, dna2, 2)

print(f"GC content DNA1: {gc1:<8} GC content DNA2: {gc2:<8}")
print(f"DNA1GC content is {result} than DNA2")
```

```python
def average_gene_expression(*args):
    sum_expression = 0
    count = 0

    for expression_value in args:
        sum_expression += expression_value
        count += 1

    average_expression = sum_expression / count
    return round(average_expression, 2)
```

```python
avg1 = average_gene_expression(120, 140)
print(avg1)
```

```python
avg2 = average_gene_expression(120, 140, 125)
print(avg2)
```

```python
avg3 = average_gene_expression(123, 91, 234, 56, 78, 77, 89, 101, 110, 95)
print(avg3)
```

```python
def average_gene_expression_kw(**kwargs):
    sum_expression = 0
    count = 0

    for exp_id, expression_value in kwargs.items():
        sum_expression += expression_value
        count += 1

    average_expression = sum_expression / count
    return round(average_expression, 2)
```

```python
avg1 = average_gene_expression_kw(
    experiment1=120,
    experiment2=130
)
print(avg1)
```

```python
avg2 = average_gene_expression_kw(
    experiment1=120,
    experiment2=130,
    experiment3=150,
    experiment4=144,
    experiment5=116,
)
print(avg2)
```
