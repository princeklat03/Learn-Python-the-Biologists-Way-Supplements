```python
import pandas as pd

df1 = pd.DataFrame({
    "Gene": ["BRCA1", "TP53", "MYC"],
    "Control_Expression": [50, 120, 80],
    "Treatment_Expression": [70, 140, 60]
})

df2 = pd.DataFrame({
    "Gene": ["EGFR", "KRAS"],
    "Control_Expression": [100, 200],
    "Treatment_Expression": [150, 180]
})
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)
```

```python
df = pd.concat([df1, df2], axis=0).reset_index().drop('index', axis=1)
df
```

```python
df4 = pd.DataFrame(
    {
        'Gene': ['BRCA1', 'TP53', 'MYC', 'EGFR', 'KRAS'],
        'Expression': [50, 120, 80, 100, 200]
    }
)

df5 = pd.DataFrame(
    {
        'Gene': ['BRCA1', 'EGFR'],
        'Chromosome': [17, 7]
    }
)
```

```python
# pd.merge(df4, df5, on="Gene", how="join_type")
```

```python
df_complete_inner = pd.merge(df4, df5, on="Gene", how="inner")
df_complete_inner
```

```python
df_complete_outer = pd.merge(df4, df5, on="Gene", how="outer")
df_complete_outer
```

```python
df_complete_left = pd.merge(df4, df5, on="Gene", how="left")
df_complete_left
```

```python
df_complete_right = pd.merge(df4, df5, on="Gene", how="right")
df_complete_right
```
