```python
# Input DNA sequence
DNA = "ATCG"

# Stepwise replacement using lowercase placeholders
seq = DNA.replace("A", "t")
seq = seq.replace("T", "a")
seq = seq.replace("C", "g")
seq = seq.replace("G", "c")

# Convert to uppercase
complement = seq.upper()

# Reverse the string to obtain reverse complement
reverse_complement = complement[::-1]

print("Original DNA:                                       ", DNA)
print("Complement:                                         ", complement)
print("Reverse complement:                                 ", reverse_complement)
```

```python
# Input DNA sequence
DNA = "ATCGATCGGGCCT"

# Count G’s and C’s
count_g = DNA.count("G")
count_c = DNA.count("C")

# Total GC count
gc_count = count_g + count_c

# Total length of the DNA sequence
total_length = len(DNA)

# GC content as a percentage
gc_content = (gc_count / total_length) * 100

# Round to two decimal places
gc_content = round(gc_content, 2)
print(f"GC content of this DNA fragment is {gc_content}%")
```
