```python
# Traditional write
path = "output.txt"
file_handle = open(path, "w")     # open in write mode
file_handle.write("This is a new file created in write mode.\n")
file_handle.close()                     # must close manually
```

```python
# With open – write mode
path = "output.txt"
with open(path, "w") as file_handle:
    file_handle.write("This is a new file created in write mode.\n")
```

```python
# Traditional append
path = "output.txt"
file_handle = open(path, "a")     # open in append mode
file_handle.write("Adding another line to the file.\n")
file_handle.close()
```

```python
# With open – append mode
path = "output.txt"
with open(path, "a") as file_handle:
    file_handle.write("Adding another line to the file.\n")
```
