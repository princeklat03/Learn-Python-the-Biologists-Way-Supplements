```python
DNA_sequence = [
    ["A", "T", "C"],
    ["A", "A", "T", "T", "G"],
    ["C", "A"],
    ["G", "G", "A", "T", "T", "C"]
    ]

length_list = []

for row in DNA_sequence:
  count = 0
  for base in row:
    count = count + 1
  length_list.append(count)
print("Lengths of nested lists:", length_list)
```
