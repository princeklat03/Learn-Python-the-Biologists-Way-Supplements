```python
gene_expression = [20, 10, 2, 30, 55, 120, 150, 200]
```

```python
for expression in gene_expression:
    print(f"Gene with expression value: {expression}")
    if expression < 20:
        print("LOW expression")
        print()   # blank line for readability
    else:
        print("HIGH expression")
        print()   # blank line for readability
```

```python
for expression in gene_expression:
    if expression < 20:
        print("Low expression")
    elif expression >= 20 and expression < 100:
        print("Moderate expression")
    else:
        print("High expression")
```

```python
sequence = "ATGCGTACGTAG"
start_codon = "ATG"

if sequence.startswith(start_codon):
    if len(sequence) % 3 == 0:
        print(f"Sequence starts with a start codon.",
                f"Sequence is a valid coding sequence.")
    else:
        print(f"Sequence starts with a start codon,",
                f"but it is NOT a valid coding sequence.")
else:
      print(f"Sequence does NOT start with a start codon,")
```
