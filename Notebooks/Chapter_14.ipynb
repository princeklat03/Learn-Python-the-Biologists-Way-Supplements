```python
dna1 = "ATGaaatttGGC"       # Mixed uppercase (exons) and lowercase (introns)
dna2 = "ATGAAATTTGGC"    # All uppercase
```

```python
pattern = r"[A-Z]+[a-z]+[A-Z]+"
```

```python
import re

match_dna1 = re.search(pattern, dna1)
match_dna2 = re.search(pattern, dna2)

if match_dna1:
  print("Match DNA1:", match_dna1.group())
else:
  print("No match DNA1")

if match_dna2:
  print("Match DNA2:", match_dna2.group())
else:
  print("No match DNA2")
```

```python
pattern_group = r"([A-Z]+)([a-z]+)([A-Z]+)"
```

```python
match_group = re.search(pattern_group, dna1)

if match_group:
  print("Whole match:", match_group.group())               # Entire matched string
  print("Exon 1:", match_group.group(1))                       # First group
  print("Intron:", match_group.group(2))                         # Second group
  print("Exon 2:", match_group.group(3))                       # Third group
else:
  print("No match found")
```

```python
import re

# Compile a pattern for Exon-Intron-Exon
pattern_group = re.compile( r"([A-Z]+)([a-z]+)([A-Z]+)")
```

```python
dna_sequences = [
    "ATGaaatttGGC",    # Exon-Intron-Exon
    "ATGCCCGGC",       # Exon only
    "TAAggtTGG"        # Exon-Intron-Exon
]
```

```python
# Compile the pattern once
pattern_group = re.compile( r"([A-Z]+)([a-z]+)([A-Z]+)")

for i, seq in enumerate(dna_sequences, 1):
  match = pattern_group.search(seq)
  if match:
    print(f"Sequence {i} matches:")
    print("  Whole match:", match.group())
    print("  Exon 1:", match.group(1))
    print("  Intron:", match.group(2))
    print("  Exon 2:", match.group(3))
  else:
    print(f"Sequence {i} does not match the pattern")
```

```python
# Find all matches in a sequence
all_matches = pattern_group.findall("ATGaaatttGGC TAAggtTGG")
print(all_matches)
```

```python
# Iterate over matches with match objects
for match in pattern_group.finditer("ATGaaatttGGC TAAggtTGG"):
 print("Whole match:", match.group())
 print("Exon 1:", match.group(1))
 print("Intron:", match.group(2))
 print("Exon 2:", match.group(3))
```
