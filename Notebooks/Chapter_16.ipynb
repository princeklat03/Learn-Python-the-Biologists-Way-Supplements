```python
genes = ["Gene1", "Gene2", "Gene3"]
expression_before = [10, 20, 15]
expression_after = [15, 13, 25]
plt.figure(figsize=(8, 5))

plt.plot(
    genes,
    expression_before,
    linewidth=1,
    linestyle='dashed',
    color='red',
    label='Before Treatment'
    )

plt.plot(
    genes,
    expression_after,
    linewidth=1.5,
    color='black',
    label='After Treatment'
    )

plt.xlabel("Gene Name", fontsize=12, fontweight='bold')
plt.ylabel("Expression", fontsize=12, fontweight='bold')
plt.title("Gene Expression Before and After Treatment", fontsize=14, fontweight='bold')

plt.legend(loc='best')
plt.show()
```

```python
dna_sequence = "ATGCGATACGATCGGCTAGCTAGCGATGCTAGCTA"
```

```python
bases = ["A", "T", "C", "G"]
```

```python
base_frequency = [dna_sequence.count(n) for n in bases]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
plt.bar(bases, base_frequency)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(bases, base_frequency, width=0.4)
plt.show()
```

```python
plt.bar(bases, base_frequency, color="black")
```

```python
colors = ["blue", "black", "red", "green"]

plt.figure(figsize=(8, 5))
plt.bar(bases, base_frequency, color=colors, width=0.4)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
colors = ["blue", "black", "red", "green"]
plt.bar(bases, base_frequency, color=colors, width=0.4)

plt.title("Nucleotide Frequency", fontsize=12, fontweight='bold')
plt.xlabel("Nucleotide")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.barh(bases, base_frequency, color=colors, height=0.4)

plt.title("Nucleotide Frequency", fontsize=12, fontweight='bold')
plt.xlabel("Frequency")
plt.ylabel("Nucleotide")

plt.show()
```

```python
# Import the required library
import matplotlib.pyplot as plt

# Sequence length data (example dataset)
sequence_lengths = [100, 201, 150, 302, 51, 80]

# Define the figure size
plt.figure(figsize=(8, 5))

# Create the histogram
plt.hist(sequence_lengths, bins=10, color= 'red', edgecolor= 'black')

# Add title and axis labels
plt.title("Sequence Length Distribution", fontsize=12, fontweight= 'bold')
plt.xlabel("Sequence Length")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
# Import required library
import matplotlib.pyplot as plt

# Labels for different species
species_labels = ['Species A', 'Species B', 'Species C', 'Species D']

# Counts of each species
counts = [25, 10, 15, 9]

# Explode values (highlight Species D)
explode_values = [0, 0, 0, 0.2]

# Create the figure
plt.figure(figsize=(6, 6))

# Create the pie chart
plt.pie(
    counts,
    labels=species_labels,
    autopct='%1.1f%%',
    shadow=True,
    explode=explode_values
    )

# Add title
plt.title('Distribution of species in a sample')
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Gene expression values (hypothetical data)
experiment_1 = [2.1, 2.3, 2.5, 2.0, 2.4, 2.2, 2.6]
experiment_2 = [4.5, 4.7, 4.6, 4.8, 5.5, 4.9, 6.2]

# Create the figure
plt.figure(figsize=(10, 6))

# Create the box plot
plt.boxplot(
    [experiment_1, experiment_2],
    tick_labels=["Experiment 1", "Experiment 2"]
    )

# Add title and labels
plt.title("Box Plot of Gene Expression Levels", fontweight='bold')
plt.ylabel("Expression Levels", fontweight='bold')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gene names and condition labels
genes = ["Gene A", "Gene B", "Gene C", "Gene D"]
conditions = ["Condition 1", "Condition 2", "Condition 3", "Condition 4"]

# Gene expression values (hypothetical data)
expression_data = [
    [2.5, 3.0, 3.5, 4.0],
    [1.2, 1.8, 2.2, 2.9],
    [4.1, 3.8, 3.6, 3.2],
    [2.0, 2.5, 3.0, 3.8],
    ]

# Create the figure
plt.figure(figsize=(10, 6))

# Create the heat map
sns.heatmap(
    expression_data,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    xticklabels=conditions,
    yticklabels=genes
    )

# Add title and labels
plt.title("Heat Map of Gene Expression", fontweight='bold')
plt.xlabel("Conditions", fontweight='bold')
plt.ylabel("Genes", fontweight='bold')

# Display the plot
plt.show()
```
