```python
string1 = "Python"

# convert to the list of characters
list_from_string1 = list(string1)
print(list_from_string1)
print(len(list_from_string1))

# convert to tuple of characters
tuple_from_string1 = tuple(string1)
print(tuple_from_string1)
print(len(tuple_from_string1))
```

```python
# From list
recovered_from_list = "".join(list_from_string1)
print("Recovered string from list:", recovered_from_list)

# From tuple
recovered_from_tuple = "".join(tuple_from_string1)
print("Recovered string from tuple:", recovered_from_tuple)
```
